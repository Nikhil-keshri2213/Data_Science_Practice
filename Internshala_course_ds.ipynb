> OOPS ?
- A programming paradigm that oraganie code into objects.
- Objects are instances of classes, and classes define the blueprints or template for creating objects.
- This paradugm focus on modeling real-world entities as objects, making it easier to design, implements and maintain spftware system.


```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def bark(self):
        print(self.name,"Bark!")

dog1 = Dog("puppy", 5)
dog2 = Dog("Sheru", 3)

print(dog1.name, "is" ,dog1.age,"years old")
print(dog2.name, "is" ,dog2.age, "years old")

dog1.bark()
dog2.bark()
```

### Class

```python
class Car:
    def __init__ (self, make, model):
        self.make = make
        self.model = model

    def display_info(self):
        print(self.make, self.model)
```

### Encapsulation and Data Hiding
- Encapsulation is the bundling of data and methods that operate on the data
into a single unit, typically called a class.

- Data hiding is an essential part of encapsulation where the internal state of an
object is hidden from the outside world.

- In Python, this is achieved by making attributes private using the __ (double
underscore) prefix.

### Inheritance and Method Overriding
- Inheritance allows a new class (subclass/derived class) to inherit attributes and methods from an existing class (superclass/base class).
- Method overriding occurs when the subclass provides a specific implementation for a method already defined in the superclass.

```python
class vehicle:
    def display_info(self):
        print("Generic class")
    
class car(vehicle):
    def __init__ (self, make, model):
        self.make = make
        self.model = model

    def display_info(self):
        print(self.make, self.model)

generic_vehicle = vehicle()
generic_vehicle.display_info()

my_car = car("Honda","City")
my_car.display_info()
```

### Handling Missing Values
- Detect Missing Values [isna() or isnull()]
    > example: missing_values = df.isnull() # or df.isna()
- Drop Missing Values 
    > df.dropna() : Removes rows with missing values
    > df.dropna(axis = 1) : Remove columns with missing values
- Filling Missing values
    > df.fillna(value) : Fill missing values with a specific value 
- Interpolate Missing values
    > df.interpolate() : Interpolate missing values

### Handling Duplicate Values
- Detect Duplicates
    * df.duplicated() : To identify duplicate rows.
    * df.drop_duplicates() : To remove them.
- Keep first or last Occurrence
    * df.drop_duplicates(keep='first') : Keep first occurrence (default)
    * df.drop_duplicates(keep='last') : keep last occurance

### Data Imputation
Process of filling missing values with estimates to make dataset complete.

* Mean_Imputed
> df.fillna(df.mean()) : Imputation with mean
* Median_Imputed
> df.fillna(df.median()) : Imputation with median
* Mode_Imputed
> df.fillna(df.mode().iloc[0]) : mode(), returns a DataFrame, so we use iloc[0] to select the first mode.

### Data Normalization and Standardization


- #### Data Normalization :
It is a process of rescaling data to a common scale without distorting the differences in the range of value.
* > x_normalizaed = (x - x_min) / (x_max - x_min)    
* x is the respective values, a_min and x_max is the minimum and maximum value in entire data range.

- #### Data Standardization : 
It is also known as feature scale, as it's a specific type of data normalization that transform the data to have a mean of zero and a standard deviation of one.
* > x_standardized = (x - mean)/standard deviation

### Fundamental of Linear Regression
Linear regression is a statistical method used to model the relationship between a dependent variable and one or more independent variable by fitting a linear equation to observed data.

It is widely used for prediction and forecasting in various fields, including economics, finance, sociology and natural science.

#### Key Concepts and Terms:
- Variables:
    - Dependent Variable(Y): This is the variable we want to predict or explain.
    It's typically denoted as Y and lies on the vertical axis of graph.

    - Independent Variable(X): These are the variables used to predict the dependent variable. They are also called variable or features.
    They are typically denoted as X and lies on the horizontal axis of a graph.

- Linear Relationship: 
It means that as the independent variable changes, the dependent variable changes proportionally.

if X increase than Y also increase, if X decrease than Y also decrease.

- Simple Linear Regression: 
In simple linear regression only one independent variable used to predict the dependent variable.

- Multiple Linear Regression: 
In simple linear regression one or more independent variable used to predict the dependent variable.

- Regression Equation
The linear regresion model is represented by an equation of the form.
* Y = B0 + B1X1 + B2X2 + .... +E
    - where, 
        1. Y is the dependent variable
        2. X1, X2, .. are the independent variable
        3. B0, B1, B2, ... are coefficients(also called sloped) or we can say that it show the effect of each independent variable on dependent variable.

- Ordinary Least Square (OLS):
    * Linear regression uses a method called ordinary least squares to estimate the coefficients.
    * It minimizes the sum of the squared differences between the observed and predicted values.
    * The best-fit line is determined by minimizing the sum of the squared vertical distances (residuals) from each data point to the line.

### Analysing The linear regression model
1. Multiple R
    - Represents the correlation between the observed and predicted values in a
    linear regression model.
    - It indicates the strength and direction of the linear relationship between the independent and dependent variables.

2. R Square
    - Also known as the coefficient of determination, it measures the proportion of the variance in the dependent variable that is predictable from the independent variables.
    - It ranges from 0 to 1, where 1 indicates a perfect fit.

3. Adjusted R Square
    - Similar to R Square but adjusted for the number of predictors in the model.
    - It penalizes the addition of unnecessary predictors, providing a more
    accurate assessment of the model's goodness of fit.

4. Standard Error
    - Represents the average deviation of the observed values from the predicted values. It indicates the accuracy of the regression model's predictions.
    - A lower standard error suggests a better fit of the model to the data.

5. Mean Square Error(MSE)
    - Mean Squared Error (MSE) is a measure of the average squared differences between predicted and actual values. A high
    - MSE indicates that the predictions are far from the actual values, suggesting that the model's performance is poor.

#### Classification Model (Logistic Regression)
- What does Logistic Regression Do ?
    - Predicts the probability that something belongs to one category or another.
    - Typically, it's used for binary classification, where there are only two possible outcomes, like "yes or no", "true or false" or "positive or negative".

- How does Logistic Regression work?
    - Works by taking a set of input features (variables or data points)
    - Applying a mathematical formula to compute a probability that the outcome is one of the two possible categories.
    - This probability can then be used to classify the outcome.
    - If the probability is above a certain threshold (usually 0.5), the prediction is "yes" or "positive"; otherwise, it's "no" or "negative."
    - The sigmoid function is used in logistic regression to map any real-valued number to a value between 0 and 1. This helps in interpreting the output as a probability.
