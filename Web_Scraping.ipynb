#### Types of web scraping:
1. Request /API scraping:
    - Directly load data from internet via "API". Considered to be the easiest type of scraping.
2. Static HTML scraping:
    - We load a simple "static" website. We extract HTML data Available on the website. Also considered to be easy.
3. Dynamic JS scraping:
    - We load a complex "dynamic" website. We interact with it like a person. Considerably more complex

### API Scraping
What is an API? 
- Api(Application Programming Interface) is a way to get an extranal program to do something.

For scraping, we send a "request" to the API program to get data for us.

What is a JSON ?
- JSON is a way to store data in text, simliar to CSV.
Very similar to python dictionaries and lists.

Dicts are called objects & List are called arrays.

We will be getting historical data for all Mutual Funds in India, from 'www.mfapi.in'

How will we be scraping?
- We will be using 'requests' python library for very easy API scraping.
- Requests will allow us to easily send requests to the API.
