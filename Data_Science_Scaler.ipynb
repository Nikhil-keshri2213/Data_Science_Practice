## Numpy: Matrix Multiplication

### Without using numpy first

```python
marks_array
```

```python
weights=[[0.4,0.2,0.2,0.2]]
```

```python
marks
```

```python
result = []
for i in range(len(marks)):
  for j in range(len(weights)):
    #marks[i] * weights[j]
    curr_sum = 0
    for k in range(len(marks[i])):
      curr_sum += marks[i][k] * weights[j][k]
    result.append(curr_sum)
```

```python
result
```

## OpenCV : Image Manipulation

```python
image_hsv = cv2.cvtColor(image_jpg, cv2.COLOR_BGR2HSV)
image_hsv.shape
```

```python
cv2.imshow('image',image_hsv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
hue, sat, val = cv2.split(image_hsv)
```

```python
add_val = 50
```

```python
val + 50
```

```python
new_val = val.astype(np.int64)+add_val
new_val[new_val>255] = 255
new_val_unint8 = new_val.astype(np.uint8)
new_val_unint8
```

```python
image_new_hsv = cv2.merge([hue,sat,new_val_unint8])
```

```python
image_new_bgr = cv2.cvtColor(image_new_hsv, cv2.COLOR_HSV2RGB)
plt.imshow(image_new_bgr)
```

### Function for change brightness of any image

```python
def change_brightness(image,add_val):
    image_hsv = cv2.cvtColor(image_jpg, cv2.COLOR_BGR2HSV)
    hue, sat, val = cv2.split(image_hsv)

    new_val = val.astype(np.int64)+add_val
    new_val[new_val>255] = 255
    new_val[new_val<0] = 0
    new_val_unint8 = new_val.astype(np.uint8)

    image_new_hsv = cv2.merge([hue,sat,new_val_unint8])
    image_new_bgr = cv2.cvtColor(image_new_hsv, cv2.COLOR_HSV2RGB)
    return image_new_bgr
```

```python
change_brightness(image_jpg,1000)
```
