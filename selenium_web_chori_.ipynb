# Web Scrapping

```python
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
```

```python
ch = webdriver.Chrome()
link = "https://www.amazon.in//s?rh=n%3A1389401031&fs=true&ref=lp_1389401031_sar"
ch.get(link)
```

```python
soup = BeautifulSoup(ch.page_source,"html.parser")
data = []
for name in soup.find_all("div",class_="a-section a-spacing-small puis-padding-left-small puis-padding-right-small"):
    namedata=(name.find("span",class_="a-size-base-plus a-color-base a-text-normal").text[0:])
    pricedata=(name.find("span",class_="a-price-whole").text[0:])
    data.append([namedata,pricedata])
    
df = pd.DataFrame(data,columns=["name","price"])
df
```
